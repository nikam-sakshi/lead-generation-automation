#STEP 1: IMPORT LIBRARIES

We start by importing required Python libraries for data processing, API calls, and Excel export.

```python
import pandas as pd
```

#STEP 2: LOAD DATASET

We load the dataset uploaded in Google Colab. This dataset contains raw lead information which will be processed further.

```python
df = pd.read_excel("/content/online_shopping_leads.xlsx")
df.head()
```

#STEP 3: EXTRACT REQUIRED COLUMNS

We extract only the required fields as per assignment:
**Name, Email, Website/LinkedIn, and Location**.

```python
df = df[["Name", "Email", "Website", "Location"]]
df.head()
```

#STEP 4: DATA CLEANING

We clean the dataset by removing duplicates and handling missing values to ensure data consistency.

```python
# Remove duplicates
df.drop_duplicates(inplace=True)

# Handle missing values
df["Email"] = df["Email"].fillna("Not Available")
df["Website"] = df["Website"].fillna("Not Available")
df["Location"] = df["Location"].fillna("Not Available")

df.head()
```

#STEP 5: EMAIL FORMAT GENERATION

We generate professional email formats automatically using the Name column to simulate business lead enrichment.

```python
df["Generated_Email"] = (
    df["Name"]
    .str.lower()
    .str.replace(" ", ".") + "@company.com"
)

df.head()
```

#STEP 6: AUTOMATION TRIGGER

We simulate automation by creating a function that runs the entire pipeline and exports data automatically.

```python
def automation_job():
    print("Running Lead Generation Automation...")

    df.to_excel("final_leads.xlsx", index=False)

    print("Automation Completed Successfully!")

automation_job()
```

#STEP 7: EXPORT TO EXCEL

We export the final cleaned and processed dataset into an Excel file for submission.

```python
df.to_excel("final_leads.xlsx", index=False)
print("Excel file created successfully!")
```
